# Machine Learning

### Aplicar clustering em profissionais da área de dados com abase em experiência e stack técnica (linguagens e cloud).
- A clusterização foi feita com base em variáveis técnicas, como anos de experiência em TI  e Dados, linguagens (Python, SQL, R etc) e nuvens (AWS, GCP, Azure).
- O uso de PCA permite reduzir de 10 variávies para 2 componentes princiapis para visualização.
- Foi escolhido n_clusters=4 foi torna adequada a segmentação de perfis distintos. E o cluster_tecnico marca a segmentação de perfis técnicos.

```python
# @title Clusterização de perfis técnicos com KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Mapeamento de faixas de tempo para valores aproximados em anos
mapa_experiencia = {
    "Menos de 1 ano": 0.5,
    "de 1 a 2 anos": 1.5,
    "de 2 a 3 anos": 2.5,
    "de 3 a 4 anos": 3.5,
    "de 4 a 5 anos": 4.5,
    "de 5 a 6 anos": 5.5,
    "de 6 a 7 anos": 6.5,
    "de 7 a 8 anos": 7.5,
    "de 8 a 9 anos": 8.5,
    "de 9 a 10 anos": 9.5,
    "Mais de 10 anos": 12
}

# Aplica o mapeamento
df["experiencia_dados_anos"] = df["2.i_tempo_de_experiencia_em_dados"].map(mapa_experiencia)
df["experiencia_ti_anos"] = df["2.j_tempo_de_experiencia_em_ti"].map(mapa_experiencia)

# Colunas para ML
colunas_ml = [
    "experiencia_dados_anos",
    "experiencia_ti_anos",
    "4.d.1_SQL",
    "4.d.2_R",
    "4.d.3_Python",
    "4.d.6_Java",
    "4.d.14_JavaScript",
    "4.h.1_Amazon Web Services (AWS)",
    "4.h.2_Google Cloud (GCP)",
    "4.h.3_Azure (Microsoft)"
]

# Prepara o DataFrame para clustering
df_ml = df[colunas_ml].copy().fillna(0)

# Normaliza os dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_ml)

# Redução de dimensionalidade para visualização
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Aplica KMeans
kmeans = KMeans(n_clusters=4, random_state=42)
df["cluster_tecnico"] = kmeans.fit_predict(X_scaled)

```

### Interpretação dos componentes principais (PCA).
- Componente 1 (Eixo X) está associado a linguagens e expêriencia técnica, sendo influenciado fortemente por: SQL, Python, Java, JavaScript, e AWS. Esse eixo pode ser interpretado como uma medida de senioridade e stack técnica mais robusta. 
- Componente 2 (Eixo Y) está ligado a linguagens alternativas e diversidade na stack, destaques em: JavaScript, Java, e Azure. Pode ser iterpretada como um indicador de stack mais variada e perfil híbrido que é "menos tradicional" do posto de vista técnico.

```python
# Interpretação das Componentes Principais (eixos X e Y)
import pandas as pd

# Cria DataFrame com os pesos (loadings) de cada variável nas componentes
componentes = pd.DataFrame(
    pca.components_,
    columns=colunas_ml,
    index=["Componente 1", "Componente 2"]
).T

# Ordena por importância absoluta em cada componente
componentes["abs_comp1"] = componentes["Componente 1"].abs()
componentes["abs_comp2"] = componentes["Componente 2"].abs()

# Mostra as variáveis que mais influenciam cada eixo
print("🔹 Top variáveis que definem o EIXO X (Componente Principal 1):")
display(componentes.sort_values("abs_comp1", ascending=False)[["Componente 1"]].head(5))

print("🔸 Top variáveis que definem o EIXO Y (Componente Principal 2):")
display(componentes.sort_values("abs_comp2", ascending=False)[["Componente 2"]].head(5))
```

```python
# @title Perfil Dominante por Cluster Técnico
# Funções auxiliares
import pandas as pd
def mais_comum(df_col):
    return df_col.value_counts().idxmax() if not df_col.isnull().all() else "N/A"

# Análise
resumo_clusters = []

for cluster in sorted(df["cluster_tecnico"].unique()):
    df_cluster = df[df["cluster_tecnico"] == cluster]

    cargo_mais_comum = mais_comum(df_cluster["2.f_cargo_atual"])
    cloud_mais_usada = mais_comum(
        df_cluster[[
            "4.h.1_Amazon Web Services (AWS)",
            "4.h.2_Google Cloud (GCP)",
            "4.h.3_Azure (Microsoft)"
        ]].idxmax(axis=1)
    )
    linguagem_mais_usada = mais_comum(
        df_cluster[[
            "4.d.1_SQL",
            "4.d.2_R",
            "4.d.3_Python",
            "4.d.6_Java",
            "4.d.14_JavaScript"
        ]].idxmax(axis=1)
    )

    resumo_clusters.append({
        "Cluster": cluster,
        "Cargo mais comum": cargo_mais_comum,
        "Cloud dominante": cloud_mais_usada.split("_")[-1],
        "Stack dominante": linguagem_mais_usada.split("_")[-1],
        "Quantidade": df_cluster.shape[0]
    })

# Exibir resultado
pd.DataFrame(resumo_clusters)
```

### Avaliação de Modelo Supervisionado.
- A modelagem usou RandomForestClassifier com 5-fold cross-validation.
- O modelo alcançou F1-socore médio de 0.696 com desvio de apenas de 0.02, o que indica consistência e estabilidade entre os folds o que sugere que há padraão nos dados capaz de prever a variável "satisfação atual".
- A variável alvo (2.k_satisfeito_atualmente) foi convertida para 0 (insatisfeito) e 1 (satisfeito), o que possiblita um problema binário claro.

```python
#  Importações
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import numpy as np

#  Prepara os dados
df_ml = df[[
    "2.k_satisfeito_atualmente", 
    "experiencia_dados_anos", 
    "experiencia_ti_anos", 
    "4.d.1_SQL", "4.d.2_R", "4.d.3_Python", "4.d.6_Java", "4.d.14_JavaScript",
    "4.h.1_Amazon Web Services (AWS)", "4.h.2_Google Cloud (GCP)", "4.h.3_Azure (Microsoft)"
]].dropna()

print(f" Total de registros válidos para modelagem: {len(df_ml)}")
if len(df_ml) == 0:
    raise ValueError("❌ Nenhum dado válido para modelar. Verifique os dados!")

#  Variável alvo
y = df_ml["2.k_satisfeito_atualmente"].astype(bool).astype(int)

#  Features
X = df_ml.drop(columns=["2.k_satisfeito_atualmente"])

#  Pipeline com normalização e modelo
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("modelo", RandomForestClassifier(n_estimators=200, class_weight="balanced", random_state=42))
])

#  Cross-validation
from sklearn.model_selection import cross_val_score

scores = cross_val_score(
    pipeline,
    X, y,
    cv=5,  # 5 folds
    scoring="f1",  # Pode trocar para "accuracy", "recall", "roc_auc", etc.
    n_jobs=-1
)

#  Resultados
print(" F1-score por fold:", np.round(scores, 3))
print(" F1-score médio:", round(scores.mean(), 3))
print(" Desvio padrão:", round(scores.std(), 3))


```
