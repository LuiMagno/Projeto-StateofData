### - 1.1 Sobre as Dados (Linhas e Colunas)

🎯 Grupo 5 — Objetivos e Processos Seletivos

🧱 Grupo 6 — Engenheiros de Dados
6.a a 6.h
(rotina, ferramentas ETL, arquitetura, data lake/warehouse, qualidade de dados)

📊 Grupo 7 — Analistas de Dados
7.a a 7.d
(rotina, ferramentas, BI self-service, cargo)

🧠 Grupo 8 — Cientistas de Dados
8.a a 8.d
(rotina, técnicas, ferramentas, uso de LLMs)

## 2.0 Selecionando Colunas - Contexto do Recomendador de Profissionais de Dados

```python
# Criando uma cópia de segurança dos dados
df = data_root.copy()
```

```python
# Dicionário temático com suas seleções de colunas
colunas_por_grupo = {
    "metadados": [col for col in df.columns if col.startswith('0.a')],
    "info_pessoal": [
        '1.b_genero', '1.c_cor/raca/etnia', '1.d_pcd', '1.g_vive_no_brasil',
        '1.h_pais_onde_mora', '1.i_estado_onde_mora', '1.l_nivel_de_ensino', '1.m_área_de_formação'
    ],
    "perfil_profissional": [
        '2.a_situação_de_trabalho', '2.d_atua_como_gestor', '2.e_cargo_como_gestor',
        '2.f_cargo_atual', '2.g_nivel', '2.h_faixa_salarial',
        '2.i_tempo_de_experiencia_em_dados', '2.j_tempo_de_experiencia_em_ti',
        '2.s_modelo_de_trabalho_ideal', '2.t_atitude_em_caso_de_retorno_presencial', '2.r_modelo_de_trabalho_atual'
    ],
    "gestor": ['3.c_responsabilidades_como_gestor', '3.d_desafios_como_gestor'],
    "funcao": ['4.a_funcao_de_atuacao', '4.a.1_atuacao_em_dados'],
    "ferramentas": [col for col in df.columns if col.startswith('4.b')],
    "fontes": [col for col in df.columns if col.startswith('4.c')],
    "linguagens": [col for col in df.columns if col.startswith('4.d')],
    "bancos": [col for col in df.columns if col.startswith('4.g')],
    "cloud": [col for col in df.columns if col.startswith('4.h')],
    "ia_generativa": [col for col in df.columns if col.startswith('4.l')],
    "copilot": [col for col in df.columns if col.startswith('4.m')],
    "oportunidade": ['5.b_oportunidade_buscada'],

    # Engenharia
    "eng_atividades": [col for col in df.columns if col.startswith('6.a')],
    "eng_etl": [col for col in df.columns if col.startswith('6.b')],
    "eng_datalake": [col for col in df.columns if col.startswith('6.d')],
    "eng_datawarehouse": [col for col in df.columns if col.startswith('6.f')],
    "eng_qualidade": [col for col in df.columns if col.startswith('6.g')],

    # Análise
    "ana_atividades": [col for col in df.columns if col.startswith('7.a')],
    "ana_ferramentas": [col for col in df.columns if col.startswith('7.b')],
    "ana_autonomia": [col for col in df.columns if col.startswith('7.c')],

    # Ciência
    "ds_atividades": [col for col in df.columns if col.startswith('8.a')],
    "ds_tecnicas": [col for col in df.columns if col.startswith('8.b')],
    "ds_tecnologias": [col for col in df.columns if col.startswith('8.c')],
}

# Juntando todas as colunas desejadas:
colunas_selecionadas = sum(colunas_por_grupo.values(), [])

# Criar o dataframe reduzido
df_reduzido = df[colunas_selecionadas].copy()

df_reduzido.head(5)
```

```python
# Tratando nulos nas colunas binárias
colunas_binarias = df_reduzido.select_dtypes(include='float').columns

df_reduzido[colunas_binarias] = df_reduzido[colunas_binarias].fillna(0.0)

# Normalizando strings
colunas_texto = df_reduzido.select_dtypes(include='object').columns

for col in colunas_texto:
    df_reduzido[col] = df_reduzido[col].astype(str).str.strip().str.lower()
```

```python
# Transformando (vetorizando) os dados para a biblioteca de recomendação
# Todas colunas binárias
colunas_binarias = df_reduzido.select_dtypes(include='float').columns.tolist()

# Todas colunas categóricas 
colunas_categoricas = df_reduzido.select_dtypes(include='object').columns.tolist()

# Todas as colunas booleanas
colunas_booleanas = df_reduzido.select_dtypes(include='boolean').columns.tolist()

# Outras colunas importantes
outras_colunas = ['0.a_token']
# Fazendo a retirada de colunas categóricas mãe que não precisam passar pelo one hot encoding

colunas_redundantes = [
    '0.a_token',
    '3.c_responsabilidades_como_gestor', 
    '3.d_desafios_como_gestor',
    '4.b_fontes_de_dados_(dia_a_dia)',
    '4.c_fonte_de_dado_mais_usada',
    '4.d_linguagem_de_programacao_(dia_a_dia)',
    '4.g_banco_de_dados_(dia_a_dia)',
    '4.h_cloud_(dia_a_dia)',
    '4.l_tipo_de_uso_de_ai_generativa_e_llm_na_empresa',
    '4.l_tipo_de_uso_de_ai_generativa_e_llm_na_empresa',
    '6.a_rotina_como_de',
    '6.b_ferramentas_etl_de',
    '7.a_rotina_como_da',
    '7.b_ferramentas_etl_da',
    '7.c_ferramentas_autonomia_area_de_negocios',
    '8.a_rotina_como_ds',
    '8.b_tecnicas_e_metodos_ds',
    '8.c_tecnologias_ds'
]

colunas_categoricas_filtradas = [col for col in colunas_categoricas if col not in colunas_redundantes]

```

```python
# 2. Criar dummies (variáveis binárias para cada categoria)
df_dummies = pd.get_dummies(df_reduzido[colunas_categoricas_filtradas], prefix=colunas_categoricas_filtradas)

df_dummies
```
